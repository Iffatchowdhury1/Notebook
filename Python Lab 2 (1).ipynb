## Euler's Method for Approximating Solutions to First-Order DEs

Consider the following initial value problem:
$$y'=f(t,y) \text{ with } y(t_0) = y_0$$

Euler's method uses succesive short line segments with slopes determined by $f(t,y)$ to trace out the approximate solution to this initial value problem.

Consider the specific example with $y' = t^2 - y$ with initial condition $y(0) = 2$. Suppose we want to use $n = 8$ steps to approximate the solution, and in particular, $y(2.5)$. We would have the following implementation of Euler's Method:

```python
def f(t,y):  
    return t**2 - y  # If you have functions like sine or cosine or exp in this function use np instead of sp.

plt.figure(figsize=(8,8))
slopefieldplot(f, -0.2, 2.8, 0.5, 3.5, 0.1, lw=2)

y = 2.0                   # This is the initial value of y
t = 0.0                   # This is the initial value of t
tfinal = 2.5              # This is the value of t we want to get to
n = 8                     # This is how many steps we want to take
delta_t = (tfinal-t)/n    # This is our step size

tlist = [t]               # Initialize lists in which to store the data
ylist = [y]               # for later plotting

for i in range(n):        # Iterate of our steps
    slope = f(t,y)        # Compute the slope at the current location
    y = y + slope*delta_t # Compute the new y value
    t = t + delta_t       # Advance in time by one step
    tlist.append(t)       # Add the new t and y values at the ends of the lists
    ylist.append(y)
    
for t,y in zip(tlist,ylist): #This just prints out the t and y values (IGNORE THIS)
    print(f'{t:8.4f} {y:12.8f}')
    
plt.plot(tlist,ylist, 'mo-', lw=3, alpha=0.6, label='Euler Approximation with %g steps' %n)
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.legend(loc='center')
```

Now, if we plan on using Euler's method over and over again, it makes sense to define a short function/program which takes in the relevant funtion, initial condition, number of steps, and final t value, and then outputs the lists of $t$ and $y$ values. 

```python
def euler_method(f,t,y,n,t_final):
    delta_t = (t_final-t)/n    # This is our step size

    tlist = [t]               # Initialize lists in which to store the data
    ylist = [y]               # for later plotting

    for i in range(n):        # Iterate of our steps
        slope = f(t,y)        # Compute the slope at the current location
        y = y + slope*delta_t # Compute the new y value
        t = t + delta_t       # Advance in time by one step
        tlist.append(t)       # Add the new t and y values at the ends of the lists
        ylist.append(y)
        
    return tlist,ylist
```

We can use this program like in the following example for the Euler approximation to the IVP
$$y'=\sin(t)y \text{ with } y(0) = 1 \text{ and } t_{final} = 2$$
where we would like to plot the approximation for multiple $n$ values on the same graph.

```python
def f(t,y):
    return np.sin(t)*y
n = 4
tlist,ylist = euler_method(f,0,1,n,2)

n1 = 8
tlist1,ylist1 = euler_method(f,0,1,n1,2)

n2 = 16
tlist2,ylist2 = euler_method(f,0,1,n2,2)

plt.figure(figsize=(8,8))
slopefieldplot(f, -0.2, 2.2, 0.5, 4, 0.1, lw=2)

plt.plot(tlist,ylist, 'ro-', lw=3, alpha=0.6, label='Euler Approximation with %g steps' %n)
plt.plot(tlist1,ylist1, 'bo-', lw=3, alpha=0.6, label='Euler Approximation with %g steps' %n1)
plt.plot(tlist2,ylist2, 'go-', lw=3, alpha=0.6, label='Euler Approximation with %g steps' %n2)
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.legend(loc='upper left')
```

### Question 4

A better approximation scheme for first order differential equations is the Runge-Kutta method. Consider an IVP
$$y' = f(t,y) \text{ with } y(t_0) = y_0$$
and a step size $h = \frac{t_{final} - t_0}{n}$ where $n$ is the number of steps taken. Everything is the same as in Euler's method, except for the computation of $y_{i+1}$. Define the following:

$$k_1 = f\left(t_i,y_i\right)$$

$$k_2 = f\left(t_i + \frac{h}{2}, y_i + \frac{k_1h}{2}\right)$$

$$k_3 = f\left(t_i + \frac{h}{2}, y_i + \frac{k_2h}{2}\right)$$

$$k_4 = f\left(t_i + h, y_i + k_3h\right)$$

Then 

$$t_{i+1} = t_i + h \text{ and } y_{i+1} = y_i + \frac{k_1 + 2k_2 + 2k_3 + k_4}{6}h$$

The following program implements the Ruge-Kutta method:

```python
def runge(f,t,y,n,t_final):
    
    h = (t_final-t)/n    # This is our step size

    tlist = [t]               # Initialize lists in which to store the data
    ylist = [y]               # for later plotting

    for i in range(n):        # Iterate over our steps
        k_1 = f(t,y)       
        k_2 = f(t+h/2,y+k_1*h/2)
        k_3 = f(t+h/2,y+k_2*h/2)
        k_4 = f(t+h,y+k_3*h)
        
        t = t + h
        y = y + (k_1+2*k_2+2*k_3+k_4)*h/6
        
        tlist.append(t)       # Add the new t and y values at the ends of the lists
        ylist.append(y)
        
    return tlist,ylist
```

Plot the Euler and Runge-Kutta approximations with $n=6$ and $t_{final}=1$ and the exact solution all on the same graph. 

**Solution**:

```python
def f(t,y):
    return y*(t**2)
n = 6
tlist,ylist = euler_method(f,0,1,n,1)

tlist1 = [0.0000, 0.1667, 0.3333, 0.5000, 0.6667, 0.8333, 1.0000]
ylist1 = [1, 1.00154, 1.01241, 1.04254, 1.10382, 1.21273, 1.39561]

tlist2,ylist2 = runge(f,0,1,n,1)

plt.figure(figsize=(8,8))
slopefieldplot(f, -0.1, 1.1, 0.8, 1.6, 0.02, lw=2)

plt.plot(tlist,ylist, 'ro-', lw=3, alpha=0.6, label='Euler Approximation with %g steps' %n)
plt.plot(tlist1,ylist1, 'go-', lw=3, alpha=0.6, label='Exact Solution with %g steps' %n)
plt.plot(tlist2,ylist2, 'yo-', lw=3, alpha=0.6, label='Runge-Kutta method with %g steps' %n)


plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.legend(loc='upper left')


```
